```python
dfP = pd.read_json('/kaggle/input/india-electoral-bonds-details-as-json-array/bonds-purcahsed-array.json')
# pd.set_option('display.float_format', lambda x: '%.0f' % x)
# dfP['ReferenceNoURN'] = dfP['ReferenceNoURN'].astype(int)
# dfP['ReferenceNoURN'] = dfP['ReferenceNoURN'].astype(str)
dfP.head()
```

```python
dfR = pd.read_json('/kaggle/input/india-electoral-bonds-details-as-json-array/bonds-redemption-array.json')
# pd.set_option('display.float_format', lambda x: '%.0f' % x)
dfR.head()
```

```python
merged_df = pd.merge(dfP, dfR, on='BondNumber', how='outer')
merged_df.head()
```

```python
merged_df.fillna(0, inplace=True)
merged_df['IssueBranchCode'] = merged_df['IssueBranchCode'].astype(int)
merged_df['SrNoP'] = merged_df['SrNoP'].astype(int)
merged_df['SrNoR'] = merged_df['SrNoR'].astype(int)
merged_df['PayBranchCode'] = merged_df['PayBranchCode'].astype(int)
merged_df['PayTeller'] = merged_df['PayTeller'].astype(int)
merged_df['IssueTeller'] = merged_df['IssueTeller'].astype(int)
merged_df['DenominationsP'] = merged_df['DenominationsP'].astype(int)
merged_df['DenominationsR'] = merged_df['DenominationsR'].astype(int)

merged_df.head()
```

```python
merged_df = merged_df.replace({0.0: ''})
merged_df.head()
```

```python
# Convert 'ReferenceNoURN' column to string without scientific notation
# merged_df['ReferenceNoURN'] = merged_df['ReferenceNoURN'].apply(lambda x: '{:.0f}'.format(x))

# Save DataFrame to JSON
merged_df.to_json('bonds-purchased-redemption-array.json', orient='records')
```

```python
# Group by 'NameOfThePoliticalParty' and sum 'DenominationsR'
summaryP = merged_df.groupby('NameOfThePurchaser')['DenominationsP'].sum().reset_index()

# Rename columns
summaryP.rename(columns={'NameOfThePurchaser': 'name', 'DenominationsP': 'amount'}, inplace=True)
summaryP.to_json('purchasers.json', orient='records')

# with open('purchasers.json', 'w') as f:
#     json.dump(purch, f)
```

```python
# Group by 'NameOfThePoliticalParty' and sum 'DenominationsR'
summary = merged_df.groupby('NameOfThePoliticalParty')['DenominationsR'].sum().reset_index()

# Rename columns
summary.rename(columns={'NameOfThePoliticalParty': 'name', 'DenominationsR': 'amount'}, inplace=True)
summary.to_json('parties.json', orient='records')
# with open('parties.json', 'w') as f:
#     json.dump(parties, f)
```

```python
# Group by 'NameOfThePoliticalParty' and 'NameOfThePurchaser' and sum 'DenominationsR'
summaryPartyW = merged_df.groupby(['NameOfThePoliticalParty', 'NameOfThePurchaser'])['DenominationsR'].sum().reset_index()

# Rename columns
summaryPartyW.rename(columns={'NameOfThePoliticalParty': 'name', 'NameOfThePurchaser': 'purchaser', 'DenominationsR': 'total_amount'}, inplace=True)

summaryPartyW.to_json('party-wise.json', orient='records')
```

```python
# Group by 'NameOfThePurchaser' and 'NameOfThePoliticalParty' and sum 'DenominationsP'
summaryCompanyW = merged_df.groupby(['NameOfThePurchaser', 'NameOfThePoliticalParty'])['DenominationsP'].sum().reset_index()

# Rename columns
summaryCompanyW.rename(columns={'NameOfThePoliticalParty': 'party', 'NameOfThePurchaser': 'name', 'DenominationsP': 'total_amount'}, inplace=True)

summaryCompanyW.to_json('company-wise.json', orient='records')
```
